```python
import pandas as pd
data = []
for i in range(401):
  article = ""
  name = f'{i+1:03}' + ".txt"
  file_1 = open(name,'r',encoding= 'unicode_escape')
  headline = file_1.readline().strip()
  while True:
    tmp = file_1.readline()
    if not tmp:
      break
    article += tmp.strip() + " "
  temp = [headline,article]
  data.append(temp)
df = pd.DataFrame(data, columns = ['headline', 'text'])
%cd  /content/drive/'My Drive'/NLP_project/dataset
df.to_csv("tech.csv", index=False)
```
